```python
import pandas as pd
import numpy as np

import requests
import datetime as dt

from datetime import datetime, timedelta

import time

import os
```

```python
path_data = r'C:\Users\patri\Taipower\data\loadarea'
```

```python
def get_load_data(date):
    payload = {'date': date}

    r = requests.get('https://purbao.lass-net.org/powerLoad', params = payload)
    
    df_temp = pd.DataFrame(r.json())
    df_temp = df_temp.rename(columns={'_id':'time'})
    df_temp['date'] = date
    df_temp['time'] = pd.to_datetime(df_temp['time'], format='%H:%M:%S').dt.time.astype('str')
    df_temp['datetime'] = pd.to_datetime(df_temp['date'] + ' ' +  df_temp['time'],
                                         errors='coerce')
    df_temp = df_temp.drop(['time','date'], axis=1)
    df_temp = df_temp.melt(id_vars='datetime', var_name='area', value_name='load')
    return df_temp
```

```python
starttime = '2022/03/14'
endtime = '2022/07/01'
delta_day = 1

timePeriod_temp = starttime

while timePeriod_temp <= endtime:
    try:
        print(timePeriod_temp)
        df_load_area_temp = get_load_data(timePeriod_temp)
        df_load_area_temp.to_csv(os.path.join(path_data, 'loadarea_'+pd.to_datetime(timePeriod_temp).strftime('%Y%m%d')+'.csv'), index=False, encoding='utf-8')
        timePeriod_temp = (pd.to_datetime(timePeriod_temp)+dt.timedelta(days=delta_day)).strftime('%Y/%m/%d')
        print('success!')
    except:
        print('error')   
        time.sleep(10)
        timePeriod_temp = (pd.to_datetime(timePeriod_temp)+dt.timedelta(days=delta_day)).strftime('%Y/%m/%d')
        
    print('------------------------------------------------------')
```
